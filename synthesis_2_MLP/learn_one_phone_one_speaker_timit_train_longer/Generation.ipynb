```python
def generate_from_data( xsamples, ysamples, f, init, sigma ):  
    wave = numpy.zeros( (1,32000) )
    wave[0,0:xsamples] = init
    
    for i in range(xsamples, wave.shape[1]-ysamples, ysamples): #Generate waveform
        newy = f( wave[0:1,i-xsamples:i] )
        wave[0,i:i+ysamples] = newy+numpy.random.normal( 0, 1, (1,ysamples))*sigma
    return wave[0,:]
             
def generate(speaker, modelnum, trainingrate, coeffs, initidx ):
    fp = open( "%d/best%d-%s.pkl"%(speaker,modelnum,trainingrate ) )
    bestMLP = cPickle.load(fp)
    fp.close()
        
    X = bestMLP.get_input_space().make_batch_theano()
    Y = bestMLP.fprop(X)
    f = theano.function( [X], Y )
    
    fp = open("%d/model_spec_%d-%s.yaml"%(speaker,modelnum,trainingrate ))
    model = pylearn2.config.yaml_parse.load(fp)
    dataset = model.dataset
    fp.close()
    
    mrse = numpy.sqrt(dataset.mse( 'train', f( dataset.getset('train')[0]) ))
       
    xsamples = dataset.getset('valid')[0].shape[1]
    ysamples = dataset.getset('valid')[1].shape[1]
    
    idx = dataset.shuffle_idxs[dataset.num_train_examples+initidx]
    sigma = mrse/dataset.scaling[idx]
    init = dataset.getset('valid')[0][initidx:(initidx+1),:]
    
    gens = []
    for coeff in coeffs:
        gens.append( (generate_from_data( xsamples, ysamples, f, init, coeff*sigma )*dataset.scaling[idx]).astype(numpy.int16) )
    
    print gens
    
    return gens
    
```

```python
import audiodisplay

def html_table( tbl ):
    s = "<table>"
    maxwidth = max( map( len, tbl) )
    for row in tbl:
        if len(row)==1:
            s += "<tr><td colspan=%d>" % maxwidth
            s += row[0]
            s += "</td></tr>"
        else:
            s += "<tr><td>" + "</td><td>".join( row ) + "</td></tr>"
    s += "</table>"
    return s

coeffs = [1,0.5,0.1,0.01,0]
#coeffs = [1,0]
html = ""
for modelnum in [2,9,13,14,16,21,22,28,29]:
    for trainingrate in ["0.3","0.1"]:
        html += str(modelnum) + " " + trainingrate + ":<br>"
        tbl = [ [""] + map(lambda x: str(x), coeffs) ]
        for speaker in [104,144,169,179]:
            row = [ str(speaker) ]
            gens = generate( speaker, modelnum, trainingrate, coeffs, 0 )
            audiodisplay.Audio( data=genwave[0,:], rate=16000 )
            for coeff,wave in zip(coeffs,gens):
                row.append( audiodisplay.Audio( data=wave, rate=16000 )._repr_html_() )
            tbl.append(row)
        html += html_table(tbl)

from IPython.display import HTML
HTML(html)            
```
