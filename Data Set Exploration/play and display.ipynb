```python
# Check for gaps or overlaps between words
with_gaps = []
with_overlaps = []
for idx in range(train.number_of_recorded_sentences()):
    wrd_idcs = train.sentence_idx_to_word_idcs(idx)
    wrd_offsets = []
    for a in wrd_idcs:
        wrd_offsets.append( train.word_idx_to_offsets(a) )
    pairs = zip( wrd_offsets[:-1], wrd_offsets[1:] )
    gaps = filter( lambda (a,b): a[1]<b[0], pairs )
    overlaps = filter( lambda (a,b): a[1]>b[0], pairs )    
    if len(gaps)>0:
        with_gaps.append(idx)
    if len(overlaps)>0:
        with_overlaps.append(idx)

print "number of overlaps",len(with_overlaps)
print "numer of gaps",len(with_gaps)
```

```python
# Check for gaps or overlaps between transcription symbols
with_gaps = []
with_overlaps = []
for idx in range(train.number_of_recorded_sentences()):
    phn_idcs = train.sentence_idx_to_phoneme_idcs(idx)
    phn_offsets = []
    for a in wrd_idcs:
        wrd_offsets.append( train.phoneme_idx_to_offsets(a) )
    pairs = zip( phn_offsets[:-1], phn_offsets[1:] )
    gaps = filter( lambda (a,b): a[1]<b[0], pairs )
    overlaps = filter( lambda (a,b): a[1]>b[0], pairs )    
    if len(gaps)>0:
        with_gaps.append(idx)
    if len(overlaps)>0:
        with_overlaps.append(idx)

print "number of overlaps",len(with_overlaps)
print "numer of gaps",len(with_gaps)
```
